# 학습이 잘 되는 모델
* 사용할 레이어

    * tf.keras.layers.BatchNormalization()
    * tf.keras.layers.Activation('swish')
* 데이터

    * 보스턴 집값 예측: https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv
    * 아이리스 품종 분류: https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv

```python
import pandas as pd
import tensorflow as tf
```

```python
boston = pd.read_csv('boston.csv')

독립 = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
종속 = boston[['medv']]
print(독립.shape, 종속.shape)
```

```python

X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python

X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
model.fit(독립, 종속, epochs=1000)
```

```python
iris = pd.read_csv('iris.csv')
iris = pd.get_dummies(iris) # dummies = 원핫 인코딩

a = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
b = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
```

```python
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)  # 비율로 예측하기 위해 활성화함수 = softmax 사용
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics = 'accuracy')
```

```python
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.Activation('swish')(H)
H = tf.keras.layers.BatchNormalization()(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.Activation('swish')(H)
H = tf.keras.layers.BatchNormalization()(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.Activation('swish')(H)
H = tf.keras.layers.BatchNormalization()(H)

Y = tf.keras.layers.Dense(3, activation='softmax')(H)  # 비율로 예측하기 위해 활성화함수 = softmax 사용
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics = 'accuracy')
```

```python
model.fit(a, b, epochs=1000)
```
